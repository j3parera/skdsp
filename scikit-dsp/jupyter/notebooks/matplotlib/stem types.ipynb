```python
x = np.linspace(0, 6*np.pi, 100)
y = np.sin(x)
figure(figsize=[16, 8], facecolor='#ddeeff')  # figsize en inches, se puede cambiar dpi = 80 por defecto
subplot(2, 2, 1)
stem(x, y)
axis('tight')
subplot(2, 2, 2)
stem(x, y, linefmt='k-', markerfmt='ko', basefmt='k')
axis('tight')
subplot(2, 2, 3)
stem(x, y, linefmt='b-', markerfmt=' ', basefmt='k')
axis('tight')
subplot(2, 2, 4)
stem(x, y, linefmt='b-', markerfmt='', basefmt='k')
axis('tight')
```
