```python
%matplotlib inline
from ipywidgets import interact, fixed
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-notebook')
from scipy.fftpack import fft, fftshift
from scipy.signal import argrelmax
```

# Ventana Hamming Generalizada
$$w_{rz}[n]=1,\quad \left|n\right| \le \frac{M-1}{2}$$
$$w_{hz}[n]= w_{rz}[n] \left[ \alpha + 2\beta \cos(\omega_M n) \right]$$
$$W_{hz}(\omega)= \alpha W_{rz}(\omega) + \beta W_{rz}(\omega-\omega_M) + \beta W_{rz}(\omega + \omega_M) ,\quad\omega_M=\frac{2\pi}{M}$$
* Centrada en el origen (zero-phase)
* Longitud M
* Hay cancelación de los lóbulos laterales
* El lóbulo principal se dobla

```python
from scipy.special import diric
def genhammingplot(M=3, alpha=0.5, beta=0.25):
    N = 20
    n = np.arange(-N, N+1)
    wr = np.zeros(len(n))
    wr[np.abs(n) <= (M-1)/2] = 1
    wh = wr * (alpha + 2*beta*np.cos(2*np.pi*n/M))
    o = np.linspace(-np.pi, np.pi, 512, endpoint=False)
    x1 = alpha*diric(o, M)
    x2 = beta*diric(o-2*np.pi/M, M)
    x3 = beta*diric(o+2*np.pi/M, M) 
    x = x1 + x2 + x3
    o /= np.pi
    xticks = np.arange(-1, 1.25, 0.25)
    plt.subplots(1, 2, figsize=(14, 4))
    plt.subplot(121)
    offx = 0.75
    offy = 0.1
    lines = plt.stem(n, wh, linefmt='k-', markerfmt='ko', basefmt='k-')
    lines[2].set_xdata([n[0]-1, n[-1]+1])
    plt.axis([np.min(n)-offx, np.max(n)+offx, np.min(wh)-offy, np.max(wh)+offy])
    plt.grid(True)
    plt.xlabel(r'$n$', fontsize=16, labelpad=-4)
    plt.title(r'$w_{hz}[n], M=' + '{0}'.format(M) + r',\alpha=' + '{0}'.format(alpha) +
             r',\beta=' + '{0}'.format(beta) + r'$', fontsize=16, loc='left')
    plt.subplot(122)
    plt.plot(o, x)
    plt.plot(o, x1, 'r--')
    plt.plot(o, x2, 'r--')
    plt.plot(o, x3, 'r--')
    plt.axis([-1, 1, -0.2, 0.6])
    plt.xticks(xticks)
    ax = plt.gca()
    ax.axhline(0, color='k', linewidth=1)
    ax.axvline(0, color='k', linewidth=1)
    plt.grid(True)
    plt.title(r'$W_{hz}(\omega), M=' + '{0}'.format(M) + r',\alpha=' + '{0}'.format(alpha) +
             r',\beta=' + '{0}'.format(beta) + r'$', fontsize=16, loc='left')
    plt.xlabel(r'$\omega/\pi\,(\rm{rad})$', fontsize=16, labelpad=-4)
    plt.ylabel('Amplitud')
interact(genhammingplot, M=(3, 21, 1), alpha=fixed(0.5), beta=fixed(0.25))
```

## Ventana de Hann (o Hanning o Coseno alzado)
* $\alpha=1/2,\beta=1/4$
$$w_{hz}[n]= w_{rz}[n] \left[ \frac{1}{2} + \frac{1}{2} \cos(\omega_M n) \right] = w_{rz}[n] \cos^2(\omega_M n)$$
$$w_{hz}[n]= \cos^2(\omega_M n),\,|n| \le (M-1)/2$$

```python
def hannplot(M=3):
    N = 20
    pad = int((2*N+1-M)/2)
    n = np.r_[-N:N+1:1]
    alpha = 0.5
    beta = 0.25
    Mo2 = int((M-1)/2)
    wh = alpha + 2*beta*np.cos(2*np.pi*(np.r_[-Mo2:Mo2+1:1]/M))
    x = np.r_[[0]*pad, wh, [0]*pad]
    Nfft = 512
    Wh = fftshift(fft(x, Nfft))
    Wh /= np.max(Wh)
    aWh = np.abs(Wh)
    mWh = 20*np.log10(aWh)
    o = np.linspace(-1, 1, Nfft)
    plt.subplots(1, 2, figsize=(14, 4))
    plt.subplot(121)
    offx = 0.75
    lines = plt.stem(n, x, linefmt='k-', markerfmt='ko', basefmt='k-')
    lines[2].set_xdata([n[0]-1, n[-1]+1])
    plt.axis([-N-offx, N+offx, -0.1, 1.1])
    plt.grid(True)
    plt.xlabel(r'$n$', fontsize=16, labelpad=-4)
    plt.title(r'$w_{hz}[n], M=' + '{0}'.format(M) + r',\alpha=' + '{0}'.format(alpha) +
             r',\beta=' + '{0}'.format(beta) + r'$', fontsize=16, loc='left')
    plt.subplot(122)
    plt.plot(o, mWh)
    plt.axis([-1, 1, -80, 3])
    plt.xticks(np.r_[-1:1.25:0.25])
    plt.axhline(0, color='k', linewidth=1)
    plt.axvline(0, color='k', linewidth=1)
    maxidx = argrelmax(np.abs(aWh), mode='wrap')
    maxWh = aWh[maxidx]
    maxWh[np.argmax(maxWh)] = 0
    at1 = 20*np.log10(np.max(maxWh))
    plt.axhline(at1, color='k', linewidth=1, dashes=(8,4))
    plt.text(0.75, at1 + 2, '{0:.1f} dB'.format(at1))
    plt.grid(True)
    plt.title(r'$W_{hz}(\omega), M=' + '{0}'.format(M) + r',\alpha=' + '{0}'.format(alpha) +
             r',\beta=' + '{0}'.format(beta) + r'$', fontsize=16, loc='left')
    plt.xlabel(r'$\omega/\pi\,(\rm{rad})$', fontsize=16, labelpad=-4)
    plt.ylabel('Magnitud (dB)')
interact(hannplot, M=(5, 39, 2))
```

## Ventana de Hamming
* $\alpha=25/46\approx0.54,\beta=(1-\alpha)/2\approx0.23$

$$w_{hz}[n]= w_{rz}[n] \left[ 0.54 + 0.23 \cos(\omega_M n) \right]$$
$$w_{hz}[n]= 0.54 + 0.46 \cos(\omega_M n),\,|n|<(M-1)/2$$

```python
from scipy.signal import hamming
def hammingplot(M=3):
    N = 30
    pad = int((2*N+1-M)/2)
    n = np.r_[-N:N+1:1]
    alpha = 0.54
    beta = 0.23
    Mo2 = int((M-1)/2)
    wh = alpha + 2*beta*np.cos(2*np.pi*(np.r_[-Mo2:Mo2+1:1]/M))
    x = np.r_[[0]*pad, wh, [0]*pad]
    Nfft = 512
    Wh = fftshift(fft(x, Nfft))
    Wh /= np.max(Wh)
    aWh = np.abs(Wh)
    mWh = 20*np.log10(aWh)
    o = np.linspace(-1, 1, Nfft)
    plt.subplots(1, 2, figsize=(14, 4))
    plt.subplot(121)
    offx = 0.75
    offy = 0.1
    lines = plt.stem(n, x, linefmt='k-', markerfmt='ko', basefmt='k-')
    lines[2].set_xdata([n[0]-1, n[-1]+1])
    plt.axis([-N-offx, N+offx, -0.1, 1.1])
    plt.grid(True)
    plt.xlabel(r'$n$', fontsize=16, labelpad=-4)
    plt.title(r'$w_{hz}[n], M=' + '{0}'.format(M) + r',\alpha=' + '{0}'.format(alpha) +
             r',\beta=' + '{0}'.format(beta) + r'$', fontsize=16, loc='left')
    plt.subplot(122)
    plt.plot(o, mWh)
    plt.axis([-1, 1, -80, 3])
    plt.xticks(np.r_[-1:1.25:0.25])
    plt.axhline(0, color='k', linewidth=1)
    plt.axvline(0, color='k', linewidth=1)
    maxidx = argrelmax(np.abs(aWh), mode='wrap')
    maxWh = aWh[maxidx]
    maxWh[np.argmax(maxWh)] = 0
    at1 = 20*np.log10(np.max(maxWh))
    plt.axhline(at1, color='k', linewidth=1.5, dashes=(4,4))
    plt.text(0.75, at1 + 2, '{0:.1f} dB'.format(at1 - 20*np.log10(np.max(Wh))))
    plt.grid(True)
    plt.title(r'$W_{hz}(\omega), M=' + '{0}'.format(M) + r',\alpha=' + '{0}'.format(alpha) +
             r',\beta=' + '{0}'.format(beta) + r'$', fontsize=16, loc='left')
    plt.xlabel(r'$\omega/\pi\,(\rm{rad})$', fontsize=16, labelpad=-4)
    plt.ylabel('Magnitud (dB)')
interact(hammingplot, M=(5, 59, 2))
```
