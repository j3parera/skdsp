```python
%matplotlib inline
from ipywidgets import interact, fixed
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
import scipy.signal as sg
import os, sys
lib_path = os.path.abspath('/home/pepo/software/toolbox/pydsp/')
sys.path.append(lib_path)
```

```python
xmax = 4
Mmin = -20
Mmax = 30
m = np.r_[Mmin:Mmax+1:1]
Lmax = 10
Pmax = 10
x1 = np.r_[[0]*(abs(Mmin)), np.random.randint(-xmax, xmax+1, Lmax), [0]*(Mmax-Lmax+1)]
x2 = np.r_[[0]*(abs(Mmin)), np.random.randint(-xmax, xmax+1, Pmax), [0]*(Mmax-Lmax+1)]
def convplot(n=0, L=10, P=5):
    x3 = sg.convolve(x1, x2)
    plt.subplots(3, 1, figsize=(14, 8.4))
    
    plt.subplot(311)
    plt.stem(m, x1, linefmt='k-', markerfmt='ko', basefmt='k-')
    plt.axis([Mmin-0.5, Mmax+0.5, -xmax-0.5, xmax+0.5])
    plt.title(r'$x_1[m]$', loc='left', size=16)
    
    plt.subplot(312)
    plt.stem(m, x2, linefmt='k-', markerfmt='ko', basefmt='k-')
    plt.axis([Mmin-0.5, Mmax+0.5, -xmax-0.5, xmax+0.5])
    plt.title(r'$x_2[n-m], n=' + '{0}'.format(n) + r'$', loc='left', size=16)

    plt.subplot(313)
    plt.stem(m, x3, linefmt='k-', markerfmt='ko', basefmt='k-')
    plt.axis([Mmin-0.5, Mmax+0.5, -xmax-0.5, xmax+0.5])
```

```python
x = interact(convplot, n=(-10, 10, 1), L=(1, 10, 1), P=(1, 10, 1))
```
